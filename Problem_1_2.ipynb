```python
#problem 1-(a)
import numpy as np 
import pandas as pd
import gurobipy as grb

df=pd.read_excel("C:/Users/何慕清/Desktop/games.xlsx")
df["Date"] = pd.to_datetime(df["Date"])
df_a=df.groupby("Home")["Date"].agg(list).reset_index()
df_a.columns = ["team_i", "Home_Date"]

for _, row in df_a.iterrows():
    print(f"team {row['team_i']}：")
    formatted_dates = [d.strftime("%Y-%m-%d") for d in row["Home_Date"]]
    print("All the dates playing home:", formatted_dates)
```

```python
#problem 1-(b)
home_vs_count = df.groupby(["Home", "Visitor"]).size().reset_index(name="home_vs_j_count")
home_vs_count.columns = ["team_i", "team_j", "home_vs_j_count"]

for team_i in home_vs_count["team_i"].unique():
    team_data = home_vs_count[home_vs_count["team_i"] == team_i]
    print(f"team {team_i}：")
    for _, row in team_data.iterrows():
        print(f"play against visitor {row['team_j']} at home：{row['home_vs_j_count']} times")
```

```python
#problem 1-(c)
away_vs_count = df.groupby(["Visitor", "Home"]).size().reset_index(name="away_vs_j_count")
away_vs_count.columns = ["team_i", "team_j", "away_vs_j_count"]

for team_i in away_vs_count["team_i"].unique():
    team_data = away_vs_count[away_vs_count["team_i"] == team_i]
    print(f"team {team_i}：")
    for _, row in team_data.iterrows():
        print(f"play against {row['team_j']} away：{row['away_vs_j_count']} times")
```

```python
#problem 1-(d)
df_d=df.groupby("Visitor")["Date"].agg(list).reset_index()
df_d.columns = ["team_i", "Visitor_Date"]

for _, row in df_d.iterrows():
    print(f"team {row['team_i']}：")
    formatted_dates = [d.strftime("%Y-%m-%d") for d in row["Visitor_Date"]]
    print("All the dates playing away:", formatted_dates)
```

```python
#problem 2
Team=df_a["team_i"]
Date=df["Date"]
Team = sorted(list(set(Team)))
Date = sorted(list(set(Date)))

H_ij = dict(
    zip(zip(home_vs_count["team_i"], home_vs_count["team_j"]),  # 键：(team_i, team_j) 元组
        home_vs_count["home_vs_j_count"]                         # 值：对应的对战次数
    )
)
A_ij = dict(
    zip(zip(away_vs_count["team_i"], away_vs_count["team_j"]),  # 键：(team_i, team_j) 元组
        away_vs_count["away_vs_j_count"]                         # 值：对应的对战次数
    )
)
#print(len(A_ij))

M=grb.Model("problem_2")
x=M.addVars(Team,Team,Date, vtype=grb.GRB.BINARY,lb=0)

for i in Team:
    for d in Date:
        if d in df_a[df_a["team_i"]==i]["Home_Date"]:
            M.addConstr(grb.quicksum(x[i,j,d] for j in Team if j != i) == 1,name=f"HomeDate_{i}_{d}")
        else:
            M.addConstr(grb.quicksum(x[i,j,d] for j in Team if j != i) == 0,name=f"NoHome_{i}_{d}")

for i in Team:
    for d in Date:
        if d in df_d[df_d["team_i"]==i]["Visitor_Date"]:
            M.addConstr(grb.quicksum(x[j,i,d] for j in Team if j != i) == 1,name=f"AwayDate_{i}_{d}")
        else:
            M.addConstr(grb.quicksum(x[j,i,d] for j in Team if j != i) == 0,name=f"NoAway_{i}_{d}")
            
for i in Team:
    for j in Team:
        if i != j:
            M.addConstr(grb.quicksum(x[i,j,d] for d in Date) == H_ij.get((i,j), 0))
for i in Team:
    for j in Team:
        if i != j:
            M.addConstr(grb.quicksum(x[j,i,d] for d in Date) == A_ij.get((i,j), 0))

M.setObjective(0, grb.GRB.MINIMIZE)
M.optimize()


if M.status in [grb.GRB.Status.OPTIMAL, grb.GRB.Status.SUBOPTIMAL]:
    schedule = []
    for d in Date:
        for i in Team:
            for j in Team:
                if i != j and x[i,j,d].X > 0.5:
                    schedule.append((d, i, j))  # (date, home, away)

    print("\n---- Constructed Schedule ----")
    for d, i, j in schedule:
        print(f"Date {d}: {j} @ {i}")

else:
    print("No feasible schedule found.")
```
