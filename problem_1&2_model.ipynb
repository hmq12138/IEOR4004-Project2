```python
import numpy as np 
import pandas as pd
import gurobipy as grb

df=pd.read_excel("C:/Users/何慕清/Desktop/games.xlsx")
df["Date"] = df["Date"].astype(str)

teams = sorted(set(df["Home"]).union(set(df["Visitor"])))

home_dates = {team: [] for team in teams}
away_dates = {team: [] for team in teams}
home_i_vs_j = {team: {t: 0 for t in teams} for team in teams}
away_i_vs_j = {team: {t: 0 for t in teams} for team in teams}

#problem 1-(a)&(d)
for team in teams:
    home_dates[team] = list(df[df["Home"] == team]["Date"])
    away_dates[team] = list(df[df["Visitor"] == team]["Date"])

#problem 1-(b)
for i in teams:
    for j in teams:
        home_i_vs_j[i][j] = df[(df["Home"] == i) & (df["Visitor"] == j)].shape[0]

#problem 1-(c)
for i in teams:
    for j in teams:
        away_i_vs_j[i][j] = df[(df["Visitor"] == i) & (df["Home"] == j)].shape[0]
'''
pd.DataFrame.from_dict(home_dates, orient="index").to_csv("C:/Users/何慕清/Desktop/home_dates.csv")
pd.DataFrame.from_dict(away_dates, orient="index").to_csv("C:/Users/何慕清/Desktop/away_dates.csv")
pd.DataFrame.from_dict({i: home_i_vs_j[i] for i in teams}).to_csv("C:/Users/何慕清/Desktop/home_i_vs_j_counts.csv")
pd.DataFrame.from_dict({i: away_i_vs_j[i] for i in teams}).to_csv("C:/Users/何慕清/Desktop/away_i_vs_j_counts.csv")
'''
```

```python
print("--------result of problem 1-(c):--------")
for i in teams:
    for j in teams:
        if i !=j:
            print(f"Team {i} played against team {j} away for {away_i_vs_j[i][j]} times")
```

```python
print("--------result of problem 1-(d):--------")
for team in teams:
    print(f"All dates when {team} played away:",away_dates[team])
```

```python
#problem 2
Team=teams
Date=df["Date"].unique()
Team = sorted(list(set(Team)))
#print(Date)

M=grb.Model("problem_2")
x=M.addVars(Team,Team,Date, vtype=grb.GRB.BINARY,lb=0)

for i in Team:
    for d in Date:
        if d in home_dates[i]:
            M.addConstr(grb.quicksum(x[i,j,d] for j in Team if j != i) == 1,name=f"HomeDate_{i}_{d}")
        else:
            M.addConstr(grb.quicksum(x[i,j,d] for j in Team if j != i) == 0,name=f"NoHome_{i}_{d}")

for i in Team:
    for d in Date:
        if d in away_dates[i]:
            M.addConstr(grb.quicksum(x[j,i,d] for j in Team if j != i) == 1,name=f"AwayDate_{i}_{d}")
        else:
            M.addConstr(grb.quicksum(x[j,i,d] for j in Team if j != i) == 0,name=f"NoAway_{i}_{d}")
            
for i in Team:
    for j in Team:
        if i != j:
            M.addConstr(grb.quicksum(x[i,j,d] for d in Date) == home_i_vs_j[i][j])
for i in Team:
    for j in Team:
        if i != j:
            M.addConstr(grb.quicksum(x[j,i,d] for d in Date) == away_i_vs_j[i][j])

for i in Team:
    for d in Date:
        M.addConstr(grb.quicksum(x[i,j,d]+x[j,i,d] for j in Team if j!=i) <= 1)

M.setObjective(0, grb.GRB.MINIMIZE)
M.optimize()


if M.status in [grb.GRB.Status.OPTIMAL, grb.GRB.Status.SUBOPTIMAL]:
    schedule = []
    for d in Date:
        for i in Team:
            for j in Team:
                if i != j and x[i,j,d].X > 0.5:
                    schedule.append((d, i, j))  # (date, home, away)

    print("\n---- Constructed Schedule ----")
    df = pd.DataFrame(schedule, columns=["Date", "team playing home","team playing away"])
    print(df)
    df.to_excel("C:/Users/何慕清/Desktop/problem2.xlsx",index=False)


else:
    print("No feasible schedule found.")
```
